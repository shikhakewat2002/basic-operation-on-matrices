```python
import numpy as np
A= np.array([[1,2,3],
            [2,3,4],
            [4,5,6]])
B=np.array([[2,4,3],
           [3,3,3],
           [4,3,5]])
print("Matrix of A is :\n",A,"\n")
print("Matrix of B is :\n",B,"\n")
addition= A+B
print("\naddition of A and B:\n", addition)
subtraction=A-B
print("\nsubtraction of A and B:\n", subtraction)
matrix_multiplication= A@B
print("\nmatrix multiplication of A and B:\n",matrix_multiplication)
element_wise_multiplication= A*B
print("\nelement-wise multiplication of A and B:\n",element_wise_multiplication)
print("\n scaler multiplication of A:\n", 2*A)
```

```python
import numpy as np


rows = int(input("Enter number of rows: "))
cols = int(input("Enter number of columns: "))

print("\nEnter elements of Matrix A:")
A_list = []
for i in range(rows):
    row = list(map(int, input(f"Row {i+1}: ").split()))
    A_list.append(row)

print("\nEnter elements of Matrix B:")
B_list = []
for i in range(rows):
    row = list(map(int, input(f"Row {i+1}: ").split()))
    B_list.append(row)


A = np.array(A_list)
B = np.array(B_list)


print("\nMatrix A is:\n", A, "\n")
print("Matrix B is:\n", B, "\n")

addition = A + B
print("\nAddition of A and B:\n", addition)

subtraction = A - B
print("\nSubtraction of A and B:\n", subtraction)

matrix_multiplication = A @ B
print("\nMatrix multiplication of A and B:\n", matrix_multiplication)

element_wise_multiplication = A * B
print("\nElement-wise multiplication of A and B:\n", element_wise_multiplication)

print("\nScalar multiplication of A (2 * A):\n", 2 * A)
```

```python
import numpy as np

#Taking matrix input from the user
rows = int(input("Enter number of rows for Matrix A and B: "))
cols = int(input("Enter number of columns for Matrix A and B: "))

def read_matrix(name):
    print(f"\nEnter elements of Matrix {name}:")
    matrix = []
    for i in range(rows):
        while True:
            row_input = input(f"Row {i+1} ({cols} values, separated by commas or spaces): ")
            # Try splitting by comma first, then by space if that yields one element
            if ',' in row_input:
                row_str_list = row_input.split(',')
            else:
                row_str_list = row_input.split()

            try:
                row = list(map(int, row_str_list))
                if len(row) == cols:
                    matrix.append(row)
                    break
                else:
                    print(f"❗ Error: Please enter exactly {cols} values.")
            except ValueError:
                print("❗ Error: Please enter valid integers.")
    return np.array(matrix)

A = read_matrix("A")
B = read_matrix("B")

#Operations
print("\nMatrix A is:\n", A)
print("\nMatrix B is:\n", B)

print("\nAddition of A and B:\n", A + B)
print("\nSubtraction of A and B:\n", A - B)

# Matrix multiplication only works if cols(A) == rows(B)
if A.shape[1] == B.shape[0]:
    print("\nMatrix multiplication of A and B:\n", A @ B)
else:
    print("\n❗ Matrix multiplication not possible: Columns of A != Rows of B")

print("\nElement-wise multiplication of A and B:\n", A * B)
print("\nScalar multiplication of A (2 * A):\n", 2 * A)
```
