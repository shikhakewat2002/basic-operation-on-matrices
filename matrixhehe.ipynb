```python
import numpy as np
A= np.array([[1,2,3],
            [2,3,4],
            [4,5,6]])
B=np.array([[2,4,3],
           [3,3,3],
           [4,3,5]])
print("Matrix of A is :\n",A,"\n")
print("Matrix of B is :\n",B,"\n")
addition= A+B
print("\naddition of A and B:\n", addition)
subtraction=A-B
print("\nsubtraction of A and B:\n", subtraction)
matrix_multiplication= A@B
print("\nmatrix multiplication of A and B:\n",matrix_multiplication)
element_wise_multiplication= A*B
print("\nelement-wise multiplication of A and B:\n",element_wise_multiplication)
print("\n scaler multiplication of A:\n", 2*A)
```

```python
rows = int(input("Enter the number of rows: "))
columns = int(input("Enter the number of columns: "))

print("\nEnter elements of matrix S:")
S= np.array([[int(input(f"A[{i+1},{j+1}]:")) for j in range(columns)] for i in range(rows)])

print("\nEnter elements of matrix K:")
K= np.array([[int(input(f"B[{i+1},{j+1}]:")) for j in range(columns)] for i in range(rows)])

print("\nMatrix S is :\n",S,"\n")
print("Matrix K is :\n",K,"\n")

print("\naddition of S and K:\n", S+K)
print("\nsubtraction of S and K:\n", S-K)
print("\nElement-wise multiplication:\n",S*K)


row_num =  int(input("\nEnter row number to display of Matrix S: "))-1
col_num = int(input("Enter column number to display of Matrix K: "))-1
print(f"\nRow {row_num+1} of Matrix S is: {S[row_num]}")
print(f"column {col_num+1} of Matrix K is: \n{K[:,col_num]}")
```
